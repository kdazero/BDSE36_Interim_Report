```python
import json
import matplotlib.pyplot as plt

# 載入單人遊戲和多人遊戲的 JSON 檔案
with open('single_player_games.json', 'r', encoding='utf-8') as f:
    single_player_games = json.load(f)

with open('multi_player_games.json', 'r', encoding='utf-8') as f:
    multi_player_games = json.load(f)

# 計算單人遊戲和多人遊戲中，免費和付費遊戲的數量
def count_game_types(games):
    game_types = {
        'Free': 0,
        'Paid': 0
    }
    for game in games:
        if game['price'] == 0:
            game_types['Free'] += 1
        else:
            game_types['Paid'] += 1
    return game_types

single_player_game_types = count_game_types(single_player_games)
multi_player_game_types = count_game_types(multi_player_games)

# 繪製圖表
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# 繪製單人遊戲的圖表
axes[0].pie(single_player_game_types.values(), labels=single_player_game_types.keys(), autopct='%1.1f%%', startangle=140)
axes[0].set_title('Single-Player Games')

# 繪製多人遊戲的圖表
axes[1].pie(multi_player_game_types.values(), labels=multi_player_game_types.keys(), autopct='%1.1f%%', startangle=140)
axes[1].set_title('Multi-Player Games')

plt.suptitle('Share of free and paid games among single-player and multiplayer games', fontsize=16)
plt.subplots_adjust(top=0.5)
plt.tight_layout()
plt.show()
```
