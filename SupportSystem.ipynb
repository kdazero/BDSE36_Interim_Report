```python
import json
import matplotlib.pyplot as plt

# 載入免費和付費遊戲的 JSON 檔案
with open('free_games.json', 'r', encoding='utf-8') as f:
    free_games = json.load(f)

with open('paid_games.json', 'r', encoding='utf-8') as f:
    paid_games = json.load(f)

# 將免費和付費遊戲資料合併
all_games = free_games + paid_games

# 統計遊戲支援系統的數量
def count_support_systems(games):
    systems = {
        'Windows': 0,
        'Mac': 0,
        'Linux': 0
    }
    for game in games:
        if game['supportWindows']:
            systems['Windows'] += 1
        if game['supportMac']:
            systems['Mac'] += 1
        if game['supportLinux']:
            systems['Linux'] += 1
    return systems

system_counts = count_support_systems(all_games)

# 計算每個系統的占比
total_games = len(all_games)
system_percentages = {system: (count / total_games) * 100 for system, count in system_counts.items()}

# 繪製圖表
plt.figure(figsize=(8, 6))
plt.bar(system_percentages.keys(), system_percentages.values(), color=['#5599FF', '#BBFF66', '#FFA488'], width=0.5)
plt.xlabel('OS')
plt.ylabel('Available System (%)')
plt.title('Game Support OS Statistics')

# 在每個長條上顯示占比
for system, percentage in system_percentages.items():
    plt.text(system, percentage, f'{percentage:.1f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
